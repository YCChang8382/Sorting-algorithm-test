```python
import random
import time #時間函式庫,使用time方法確認"開始"與"結束"的時間
import sys #sys函式庫，使用getsizeof方法確認各個排序法的記憶體使用量
import itertools

# 1. 各個排列法程式碼

def bubble_sort(arr, low=None, high=None):#因為quicksort需要low&high參數，在沒有用到的排序法位置設定為None
    for i in range(len(arr)-1):#bubble sort須執行n-1次
      swap = True #假設已經排列完成
      for j in range(0, len(arr)-i-1):#由左到右進行檢測
        if arr[j]>arr[j+1]:  #比較鄰近數字大小
          arr[j],arr[j+1]=arr[j+1],arr[j] #如有排列錯誤則互換
          swap = False#排列還未完成
          if swap == True:
            break
    return arr #return在最外圈的for迴圈
    pass
def selection_sort(arr, low=None, high=None):
    for i in range(len(arr)-1):#selection sort須執行n-1次
      min_num=i
      for j in range(i+1, len(arr)):
        if arr[min_num]>arr[j]:
          min_num=j
      arr[i],arr[min_num]=arr[min_num],arr[i]
    return arr
    pass
def insertion_sort(arr, low=None, high=None):
    for i in range(1,len(arr)):#從[0]跟[1]個數字開始比較
      insert_num=arr[i]#insert_num儲存目前要插入的數字
      j=i-1#先跟i前面的一個數字比較
      while(j>=0 and insert_num<arr[j]):
        arr[j+1]=arr[j]#將arr[j]往後移一位
        j-=1#跟原本j的前面1位的數字比較，直到insert_num<=arr[j]
      arr[j+1]=insert_num
    return arr
    pass
def merge_sort(arr, low=None, high=None):
  if len(arr)==1:
    return arr
  elif len(arr)>1:
    mid=len(arr)//2 #紀錄序列拆分後的長度 需用//取整數
    l_arr=arr[:mid] #左半序列
    r_arr=arr[mid:] #右半序列

    merge_sort(l_arr)#遞迴進行排列，也可以另外寫成另一個函式
    merge_sort(r_arr)

    r_num=0
    l_num=0
    merge_num=0 #記錄三個子序列的位置

    while r_num<len(r_arr) and l_num<len(l_arr): #比較左右序列，限定比較次數為序列長度
      if r_arr[r_num]<l_arr[l_num]: #左右從第0項開始相比，如右側較小
        arr[merge_num]=r_arr[r_num] #右側值紀錄為第一個
        #pythonic寫法: merge_num.append(r_arr[r_num])
        r_num+=1
      else: #左側亦然
        arr[merge_num]=l_arr[l_num]
        l_num+=1
      merge_num+=1  #將merge值+1繼續記錄
    while r_num<len(r_arr): #右側序列內自行比較
      arr[merge_num]=r_arr[r_num]
      r_num+=1
      merge_num+=1
    while l_num<len(l_arr): #右側序列內自行比較
      arr[merge_num]=l_arr[l_num]
      l_num+=1
      merge_num+=1
  return arr
  pass
def quick_sort(arr, low, high):
  if low< high:
    pi = partition(arr, low, high)
    quick_sort(arr, low, pi-1)
    quick_sort(arr, pi+1, high)
  return arr

def partition(arr, low, high):
  pivot = arr[high]
  i=low
  j=high-1
  pivot=arr[high]

  while i<j:#如果左邊的數<右邊的數
    while i<high and arr[i]<pivot: #非最右側數字且比pivot小
      i+=1 #將i右移
    while j>low and arr[j]>pivot:
      j-=1 #將j左移
    if i<j:
      arr[i],arr[j]=arr[j],arr[i]
  if arr[i]>pivot:
      arr[i],arr[high]=arr[high],arr[i]
  return i
pass

# 2. 定義測試空間與時間的函式
def test_method(method, arr, low=None, high=None):
    start_time = time.time()
    if low is not None and high is not None:
      result = method(arr,low,high)
    else:
      result = method(arr)
    end_time = time.time()
    execution_time = end_time - start_time
    memory_usage = sys.getsizeof(result)  # 取得使用記憶體用量
    return result, execution_time, memory_usage

# 3. 主函式

def main():
    numbers = [str(random.randint(1, 100000)) for _ in range(50000)]
    str_arr = ' '.join(numbers)# 輸入測資序列
    arr=[int(num) for num in str_arr.split()]
    methods = [bubble_sort, selection_sort, insertion_sort, merge_sort, quick_sort]  # List of methods to test
    results = {}

    for method in methods:
        result, execution_time, memory_usage = test_method(method, arr, low, high)
        method_name = method.__name__
        results[method_name] = {'result': result, 'execution_time': execution_time, 'memory_usage': memory_usage}

    # 4. 印出結果
    for method_name, result_data in results.items():
        print(f"Method: {method_name}")
        print(f"Result: {result_data['result']}")
        print(f"Execution Time: {result_data['execution_time']} seconds")
        print(f"Memory Usage: {result_data['memory_usage']} bytes")
        print("\n")

if __name__ == "__main__":
    main()
```
